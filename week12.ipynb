```python
import csv
f = open('seoul.csv', 'r', encoding='cp949')
data = csv.reader(f, delimiter=',')
header = next(data)
max_gap_day = ''
max_temp_day = ''
min_temp_day = ''
max_gap = 0
max_temp = -999
min_temp = 999

i = 0
j = 0
k = 0
for row in data:
    if i == 366:
        break;
    day_max = float(row[4])
    day_min = float(row[3])
    gap = day_max-day_min
    if gap > max_gap :
        max_gap = gap
        max_gap_day = row[0]
    i=i+1
    
    if j == 366:
        break;
    temp = float(row[4])
    if temp > max_temp :
        max_temp = temp
        max_temp_day = row[0]
    j=j+1
    
    if k == 366:
        break;
    temp = float(row[3])
    if temp < min_temp :
        min_temp = temp
        min_temp_day = row[0]
    k=k+1

f.close()
print("일교차가 가장 큰 날", max_gap_day)
print("최저기온이 가장 낮은 날", min_temp_day)
print("최고기온이 가장 높은 날", max_temp_day)
```

```python
import csv
f = open('subwayfee.csv', 'r', encoding='cp949')
data = csv.reader(f, delimiter=',')
header = next(data)
max_1 = 0.0
max_2 = 0
max_3 = 0
min_1 = 9999999
min_2 = 9999999
min_3 = 9999999
station_1 = ''
station_2 = ''
station_3 = ''
station_4 = ''
station_5 = ''
station_6 = ''


for row in data:
    row[4] = row[4].replace(',','')
    people = (int(row[4]))
    
    if people < min_1:
        min_1 = people
        station_4 = row[3]
        
    elif min_1 < people < min_2:
        min_2 = people
        station_5 = row[3]
            
    elif min_2 < people < min_3:
        min_3 = people
        station_6 = row[3]
        
    if max_1 < people:
        max_1 = people
        station_1 = row[3]

    elif max_1 > people > max_2:
        max_2 = people
        station_2 = row[3]

    elif max_2 > people > max_3:
        max_3 = people
        station_3 = row[3]
            

f.close()

# 지하철 노선별 역별 이용현황
print("이용객이 가장 많은 역 순위")
print("1. ", station_1, "(", max_1, ")")
print("2. ", station_2, "(", max_2, ")")
print("3. ", station_3, "(", max_3, ")")
print("이용객이 가장 적은 역 순위")
print("1. ", station_4, "(", min_1, ")")
print("2. ", station_5, "(", min_2, ")")
print("3. ", station_6, "(", min_3, ")")
```
