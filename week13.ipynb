```python
# 실습 1
import csv
import matplotlib.pyplot as plt

f = open('seoul2.csv', 'r', encoding='cp949')
data = csv.reader(f, delimiter=',')
header = next(data)
high = []
low = []
for row in data:
    if row[0].split('-')[1] == '12' and row[0].split('-')[2] == '28':
        high.append(float(row[-1]))
        low.append(float(row[-2]))
plt.plot(high)
plt.plot(low)
plt.show()
```

```python
# 실습 2
import csv
import matplotlib.pyplot as plt

f = open('seoul2.csv', 'r', encoding='cp949')
data = csv.reader(f, delimiter=',')
header = next(data)
high = []
low = []
label = ['aug', 'jan']

for row in data:
    month = row[0].split('-')[1]
    if row[-1] != '':
        if month == '08':
            high.append(float(row[-1]))
        if month == '01':
            low.append(float(row[-1]))

plt.hist(high, bins=100, color='r', label=label[0]) # min, max 사이의 칸이 100갠
plt.hist(low, bins=100, color='b', label=label[1])
plt.show() # 완료
```
